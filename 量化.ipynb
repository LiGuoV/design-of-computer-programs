# 风险因子`risk Factor`& alpha因子`α facotors`
在risk model中,根据 风险因子对已知常见因子带来的股价变化,进行建模,这些因子也许能对各个资产价格变化做出解释,称为系统收益或系统波动性,因为它们也可能显著影响 portfolio 的波动性,但不能以收益率形式带来相应的补偿,所以我们的目标是portfolio 对风险因子的`暴露`  中性化,使剩余portfoli price变化由α解释,类似消除环境噪声,隔离对话

# α Factor
## 数据结构
`α值`:单只股在单个时间步内的单个矢量 持仓方向和 权重(即:与分配的金额成正比)
`α向量` α值列表  在每个时间步都标准化 均值为0 和为1 
`universe`:一组考虑的资产 未必持有 非静态的 注意前瞻性偏差 典型的是幸存者偏差
每个步更新 或跟踪指数名单 标普500
`α factor` α向量的时间序列 `二级索引[时段, 资产名] =>α值`
`α Model` :算法 把数据转化成交易信号

##  研究流程
将假设变成代码,并根据数据测试代码.`α factor` 是数据处理工具,告诉我们股票相对于其它股的业绩,它可能预测未来均值`收益率returns,`而风险因子提供收益率方差共性方面信息,寻找alpha就是寻找偏离有效市场假说的信号
> 有效市场假说通常指所有可用信息都体现在了价格数据上,表明资产价格比较合理.

在寻找有效市场假说的例外情况时,我们寻找的是市场错误定价,简称套利.买入定价过低资产,做空可能定价过高资产,预期收益率超过了所承担风险应该带来的收益率.

## 2 处理原始α的技巧
- 横截面`中性化neutralize`
 要让`投资组合portfolio`对公共`风险因子Risk Facotors`(市场风险,行业风险)保持中性,但不能到优化步骤才考虑,以免面目全非,因子定义阶段就要控制风险
和为0表示对市场`暴露(exposure 或称 β)`为0 
做空做多一样大小头寸,各行业做空做多一样大小头寸
>1. 重要的合理假设: 股票众多 单个股票β都是1 
>2. α向量中的 α值去均值  
>影响所有股票的一般市场变化被抵消了
>3. 对`行业Sector`风险也`中性Neutral` (做空做多权重相等)即:去(行业权重)均值 

- `排名rank`
不受离群值(妖股)和噪点(小变动)影响
由于每个时间步α向量都在变化 排名更稳健 
长度一样的α向量具有相同的范围
- `z评分 zscore`
标准化 即减去均值除以标准差 使数据保持一致的范围和分布
长度不同的α向量 rank之转换为 zscore ,能重新调整范围
进而便于组合,比较 多个的因子
`(row - row.mean()) / row.stddev()`
正常分布因子值现在是在 -2 和 +2 之间带有小数精度的值
- `平滑法 smoothing`
数据个别缺失稀疏 (基本面数据 季度财报更新频率慢)
平均数 or加权平均数填充
注意，不需要使其变成行业中性，因为原始因子已经按照行业去均值



- ####以其它因子为条件
"飙升的可能逆转" ;"低交易量的保持动量";"AND"
 因子间相乘
十分位嵌套 10X10乘法表

>这些技巧 将α中的数字转变成信号 表示买入或做空投资组合中每只股票及金额 来帮助实现 为真实投资组合设定的目标.
