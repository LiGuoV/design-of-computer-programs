## 形式(Form)

```
> 1
1
```

## 求值 (Evaluation)
一般规则 由左至右对实参求值，将它们的数值传入传入**以操作符命名的函数**
特殊规则 `特殊的操作符` 或者宏
- ``quote 缩写 ` `` 什么也不做返回 **保护表达式不被求值**
##  数据 (Data)
- 数字 
- 字符串 
- 符号  要引用 不对自身求值 转大写单词 `'btc`
- 列表 要引用，不然是函数调用 ` '(+ 1 1)`
   - 函数**list** 创建列表
   - 空列表 nil 或者()空括号
## 列表操作
构造
`cons` 元素+列表 (=>(进而)`list` 多元素加空列表)
`car`第一元素(=>third) 
`cdr` 后面列表 
## 真与假 (Truth)
谓词（predicate）函数返回真假 后缀`p`
`null not`  判断空表 反义 它俩一样
`特殊操作符 
if` (test表达式)(then表达式)(else表达式)
逻辑操作符 `and `和 `or` 是宏 绕过一般的求值规则
## 函数 (Functions)
```lisp
> (defun our-third (x)
           (third x))
OUR-THIRD
> (our-third '(a b c d))
C
```
第二个实参` (x)` ,占位符符号`x`叫做`变量`。
当`变量`代表了传入函数的`实参`时，又被叫做`形参`
`符号`是`变量的名字`
>列表必须被引用，不然会被视为代码。符号必须要被引用，不然会被当作变量。
## 递归 (Recursion) 
调用自己的函数

## 输入输出 (Input and Output)
```
> (format t "~A plus ~A equals ~A. ~%" 2 3 (+ 2 3))
2 plus 3 equals 5.
NIL
```
~A 表示模板里要被填入的位置 ~%换行
> 没有副作用，则定义多个表达式的函数主体就没意义 (返回最后 前面的舍弃)

##  赋值 (Assignment)
`setf 操作符` 参1可以是符号 表达式 或变量名
``` lisp
>(setf a 'b
      c 'd
      e 'f)
> (setf (car x) 'n)
N
```
## 函数式编程
避免使用如 setf
幂等 
允许交互式测试
## 迭代
`do 宏`基本迭代操作符
```lisp
(defun show-squares (start end)
  ;; do 可以创建变量
  (do ((i start (+ i 1))) 
;; ( (i start (+ i 1)) ,nil)  一组变量的规格说明列表 形式:(variable initial update)
      ((> i end) 'done) ; 表达式1 测试表达式是结束 
    (format t "~A ~A~%" i (* i i))))
```
`progn` 包装多个表达式 依次求值 
`dolist` (variable expression)(body)
## 函数作为对象
`特殊操作符function` 缩写`#'`  称之为升引号（sharp-quote）
```lisp
> (function +)
#<Compiled-Function + 17BA4E>
> #'+
#<Compiled-Function + 17BA4E>
```
函数当作实参传入 `apply`最后一个实参是列表 `funcall`实参不需要包装成列表
